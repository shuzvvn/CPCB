# Q1.4

##Q1.4.b

```python
data= np.loadtxt("HIVseries.csv",delimiter=',')
t_data= data[:,0]
V_data= data[:,1]
```

```python
t= np.linspace(0,8,100)
V0 = 1.061e+05

# arbitrary values
kI = 0.3
kV = 0.5
beta = 1


X = beta / (kV - kI)


# trial solution
Nv = X * np.exp(-kI * t) + (V0 - X) * np.exp(-kV * t)

plt.grid(which='both')
plt.semilogy(t_data,V_data,'ro')
plt.title('Time course of HIV infection after administering')
plt.xlabel('time (days)')
plt.ylabel('viral concentration (RNA copies/mL)')

plt.semilogy(t, Nv)

plt.legend(['Datapoints', r'$N_{v}(t)$'])
```

##1.4.c
#####Nv_asymps = X * np.exp(-kI * t)

```python
# lay a ruler, find two points on that straight line
# (t, Nv_asymp): 
# (6, 10^4)
# (1, 10^5)
# solve the equations and got value for kI and X

kI = 1 / (5 * np.log10(np.e))
X = 10**(26/5)

Nv_asymp = X * np.exp(-kI * t)

plt.grid(which='both')
plt.semilogy(t_data,V_data,'ro')
plt.title('Time course of HIV infection after administering')
plt.xlabel('time (days)')
plt.ylabel('viral concentration (RNA copies/mL)')

plt.semilogy(t, Nv_asymp)

plt.legend(['Datapoints', r'$N_{v, asymp}(t)$'])
```

##1.4.d
#####Nv = X * np.exp(-kI * t) + (V0 - X) * np.exp(-kV * t)
substitute your values of kI and X into Eq 1.4

```python
# values from 1.4.c
kI = 1 / (5 * np.log10(np.e))
X = 10**(26/5)

# V0 from the input data
V0 = 1.061e+05

# base on the constraint
kV = (kI * X) / (X - V0)

# values for all parameters
print("kI = %f, X = %f, V0 = %f, kV = %f" % (kI, X, V0, kV) )

# trial solution (eq 1.4)
Nv = X * np.exp(-kI * t) + (V0 - X) * np.exp(-kV * t)

plt.grid(which='both')
plt.semilogy(t_data,V_data,'ro')
plt.title('Time course of HIV infection after administering')
plt.xlabel('time (days)')
plt.ylabel('viral concentration (RNA copies/mL)')

plt.semilogy(t, Nv)

plt.legend(['Datapoints', r'$N_{v}(t)$'])
```
